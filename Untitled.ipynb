```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
invoices = pd.read_csv('invoices.csv').sort_values(['created_at'], ascending=1).set_index('invoice_id')
customers = pd.read_csv('customers.csv').set_index('customer_id')
data = invoices.merge(customers, left_on='customer_id', right_index=True, how='left', suffixes=['', '_cust'])
```

```python
for col in ['created_at', 'due', 'last_payment_date', 'created_at_cust']:
    data[col] = pd.to_datetime(data[col], errors='coerce')

data['days_creat_due'] = (data['created_at'] - data['due']).astype('timedelta64[D]')
data['days_last_due'] = (data['last_payment_date'] - data['due']).astype('timedelta64[D]')
```
